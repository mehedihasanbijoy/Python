```python
import pandas as pd
import numpy as np
from numpy.random import randn
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('fivethirtyeight')
sns.set(style = 'darkgrid', color_codes = True)

iris = sns.load_dataset('iris')
iris.head(3)
```

```python
corr_matrix = iris.corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, cmap = 'coolwarm', annot = True)
plt.show()
```

```python
sns.lmplot(x = 'sepal_length', y = 'petal_length', data = iris, hue = 'species', markers = ['D', 'x', 's'], palette = 'Set1')
plt.show()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data = iris, hue = 'species', markers = ['o', 's', 'd'], size = 'petal_length', sizes = (10, 100))
plt.show()
```

```python
sns.pairplot(iris, size = 5, vars = ['sepal_length', 'petal_length'], hue = 'species')
plt.show()
```

```python
sns.pairplot(iris, hue = 'species', markers = ['s', 'd', 'o'])
plt.show()
```

```python
sns.pairplot(iris, kind = 'reg', hue = 'species', markers = ['s', 'd', 'o'])
plt.show()
```

```python
sns.boxplot(data = iris, orient = 'h')
plt.show()
```
