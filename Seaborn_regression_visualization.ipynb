```python
import pandas as pd
import numpy as np
from numpy.random import randn
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('fivethirtyeight')
sns.set(style = 'darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.regplot(x = 'total_bill', y = 'tip', data = tips, ci = 100)
plt.show()
```

```python
anscombe = sns.load_dataset('anscombe')
anscombe.head()
```

```python
anscombe['dataset'].value_counts()
```

```python
sns.lmplot(x = 'x', y = 'y', data = anscombe.query("dataset == 'I'"), ci = None, scatter_kws = {'s' : 200})
plt.show()
```

```python
sns.lmplot(x = 'x', y = 'y', data = anscombe.query("dataset == 'II'"), ci = None, order = 2, scatter_kws = {'s' : 200})
plt.show()
```

```python
sns.lmplot(x = 'x', y = 'y', data = anscombe.query("dataset == 'III'"), ci = None, order = 1, robust = True, scatter_kws = {'s' : 200})
plt.show()
```
