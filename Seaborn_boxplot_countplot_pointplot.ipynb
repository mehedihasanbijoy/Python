# Boxplot

```python
import pandas as pd
import numpy as np
from numpy.random import randn
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxplot(x = 'sex', y = 'tip', data = tips, hue = 'time')
plt.show()
```

```python
sns.boxplot(x = 'sex', y = 'tip', data = tips, hue = 'time', dodge = False)
plt.show()
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.boxplot(x = 'color', y = 'price', data = diamonds.sort_values('color'))
plt.show()
```

```python
sns.boxenplot(x = 'color', y = 'price', data = diamonds.sort_values('color'))
plt.show()
```

# Countplot

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(x = 'class', data = titanic, palette = 'ch:.8')
plt.show()
```

```python
sns.countplot(x = 'class', data = titanic, hue = 'sex')
plt.show()
```

# Pointplot

```python
sns.pointplot(x = 'sex', y = 'survived', data = titanic, hue = 'class')
plt.show()
```
