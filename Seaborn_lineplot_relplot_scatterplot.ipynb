# Lineplot

```python
import pandas as pd
import numpy as np
from numpy.random import randn
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('fivethirtyeight')
sns.set(style = 'darkgrid')
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'region', style = 'event', ci = 50, marker = True, err_style = 'bars')
plt.show()
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri.query("region == 'parietal'"), hue = 'event', units = 'subject', estimator = None, lw = 1)
plt.show()
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, ci = False)
plt.show()
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', hue = 'event', data = fmri, palette = 'winter_r')
plt.show()
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', hue = 'region', style = 'event', data = fmri, palette = 'winter_r')
plt.show()
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', hue = 'event', style = 'region', data = fmri, markers = True, ci = False, dashes = False)
plt.show()
```

```python
sns.relplot(x = 'timepoint', y = 'signal', data = fmri, kind = 'line', hue = 'subject', col = 'region', row = 'event', estimator = None, ci = False)
plt.show()
```

```python
dots = sns.load_dataset('dots')
dots.head()
```

```python
sns.lineplot(x = 'time', y = 'firing_rate', data = dots, hue = 'coherence', style = 'choice', ci = False)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lineplot(x = 'total_bill', y = 'tip', data = tips, sort = True, ci = True)
```

# Scatterplot

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', size = 'size', sizes = (10, 100))
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', style = 'day', size = 'size', sizes = (10, 100))
plt.show()
```

# Relplot

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', palette = 'winter_r')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time', palette = 'winter_r')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size')
plt.show()
```
