```python
import pandas as pd
import numpy as np
from numpy.random import randn
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('fivethirtyeight')
sns.set(style = 'darkgrid', color_codes = True)
```

# Distplot

```python
x = randn(100)

sns.distplot(x, bins = 15)
plt.show()
```

```python
sns.distplot(x, kde = False, bins = 10)
plt.show()
```

```python
sns.distplot(x, hist = False)
plt.show()
```

# Kdeplot

```python
sns.kdeplot(x)
plt.show()
```

```python
sns.kdeplot(x, shade = True)
plt.show()
```

```python
sns.kdeplot(x, bw = .2, shade = True)
plt.show()
```

# Jointplot

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
plt.show()
```

# Pairplot

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
obj = sns.PairGrid(iris)
obj.map_diag(sns.kdeplot)
obj.map_offdiag(sns.kdeplot, n_levels = 10)
plt.show()
```
